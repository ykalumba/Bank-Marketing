<font size="+3"><strong> 1. Preparing Data</strong></font>

```python
# Import libraries
import pandas as pd
```

# Import

```python
# Load CSV files into DataFrames
df1 = pd.read_csv("Bank Marketing/Bank Marketing.csv")
df2 = pd.read_csv("Bank Marketing/Bank Marketing_additional.csv")

# Print object type and shape for DataFrames
print("df1 type:", type(df1))
print("df1 shape:", df1.shape)
print()
print("df2 type:", type(df2))
print("df2 shape:", df2.shape)
```

## Clean `df1`

Inspect `df1` by looking at its [`shape`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.shape.html) attribute. Then use the [`info`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.info.htm) method to see the data types and number of missing values for each column. Finally, use the [`head`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.head.html) method to determine to look at the first five rows of the dataset.

```python
print(df1.info())
```

There is no missing data in dataframe 1.

## Clean `df2`

```python
print(df2.info())
```

```python
# Drop columns
df1.drop(columns= ['balance', 'day'], inplace = True)

# Print object type, shape, and head
print("df1 type:", type(df1))
print("df1 shape:", df1.shape)
df1.head()
```

```python
# Drop columns
df2.drop(columns= ['day_of_week', 'emp.var.rate', 'cons.price.idx','cons.conf.idx', 'euribor3m', 'nr.employed'], inplace = True)

# Print object type, shape, and head
print("df2 type:", type(df2))
print("df2 shape:", df2.shape)
df2.head()
```

There is no missing data in dataframe 2.

## Concatenate DataFrames

Use [`pd.concat`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.concat.html) to concatenate `df1`, `df2` as new DataFrame named `df`.

```python
# Concatenate df1 and df2
df = pd.concat([df1, df2])

# Print object type, shape, and head
print("df type:", type(df))
print("df shape:", df.shape)
df.head()
```

### Save data to csv

```python
df.to_csv("Bank Marketing/Bankdata.csv")
```
